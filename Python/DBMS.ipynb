# Key:-A key in DBMS is an attribute(Column) or a set of attributes(Column) that help to uniquely identify a tuple (or row) in a relation (or table).

# Primary Key:- PRIMARY KEY in DBMS is a column or group of columns in a table that uniquely identify every row in that table. OR Most appropriate keys and leave primary key all keys are known as Alternative keys.<br> <br>Alternate Key:- Alternate Key or Secondary Key is the key that has not been selected to be the primary key, but are candidate keys.

# Candidate Key:- Minumum number of column to required become a primary keys  known as Candidate Key.

# Primary Key:- {Unique+Not Null}
# Candidate Key:- {Unique}

# Foreign Key:- It is an attribute(column) or set attribute(column) that take reference from Primary key of same table or another table(Relation).

- A foreign key is the one that is used to link two tables together via the primary key.

# Superkey:- A superkey is a combination of all attributes(column) which can be uniquely identify two tuples in a table.
- In superkey atleast in one attribue(column) in a candidate key.

# Referential Integrity:- Referential Integrity Rule in DBMS is based on Primary and Foreign Key. The Rule defines that a foreign key have a matching primary key Reference from a table to another table should be valid.<br>Integrity Means same values for database.<br><br> Referential Integrity Rule example:- <br><br>The rule states that the DEPT_ID in the Employee table has a matching valid DEPT_ID in the Department table.<br><br>To allow join, the referential integrity rule states that the Primary Key and Foreign Key have same data types.<br>

# To maintain the Referential Integrity we use automatic solution called ON DELETE CASCADE OPTION:-ON DELETE CASCADE option to specify whether you want rows deleted in a child table when corresponding rows are deleted in the parent table. If you do not specify cascading deletes, the default behavior of the database server prevents you from deleting data in a table if other tables reference it.
