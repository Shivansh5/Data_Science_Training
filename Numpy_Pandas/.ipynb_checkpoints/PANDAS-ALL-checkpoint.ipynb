```python
# Creating Function
def get_match_city(city):
    mask = df["city"]==city
    return df[mask].shape[0]
```

```python
get_match_city("Bangalore")
```

```python
get_match_city("Mumbai")
```

```python
# After 2010 how many match held in Hydrabad city.
mask1 =df["date"]>"2010-01:01"
mask2 =df["city"]=="Hyderabad"
df[mask1 & mask2].shape[0]
```

```python
# Creating Function
def get_city_month(date,city):
    mask1 =df["date"]>date
    mask2 =df["city"]==city
    return df[mask1 & mask2].shape[0]
```

```python
get_city_month("2010-01:01","Hyderabad")
```
