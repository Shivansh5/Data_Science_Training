# Scatter 
## Scatter is generally used to plot numerical type data in doted format.

```python
exp=[1,5,6,4,3,9,8]
sal=[1000,8000,5000,3000,2000,8000,60000]
```

```python
plt.style.use("fivethirtyeight")
plt.scatter(exp,sal,color="red",marker="*")
plt.xlabel("years")
plt.ylabel("sal")
plt.title("Salary_Table")
plt.show()
```

# Plot 
# plot used to represent data in line format spically for numerical data 

```python
prod_id=[1,2,3,4,5,6,7,8,9]
offline_price=[123,444,558,999,125,456,789,345,879]
online_prce=[222,489,600,1200,200,550,800,450,900]
chor_id=[100,340,450,800,100,300,600,200,600]
#plt.scatter(prod_id,offline_price)
plt.plot(prod_id,offline_price,label="ofline_price")
plt.plot(prod_id,online_prce,label="online_price")
plt.plot(prod_id,chor_id,label="chor_bazar_price")
plt.xlabel("Prod_id")
plt.ylabel("Price")
plt.legend()
plt.show()
```

# *Bar* 

```python
year=[2012,2013,2014,2015,2016]
price=[65000,69000,75000,55000,52000]
plt.bar(year,price,edgecolor="red",color="yellow")
plt.xlabel("year")
plt.ylabel("price")
plt.show()
```
